```python
# define a function and then subs
import sympy as sym

x = sym.symbols('x')
y = x**2 - 3*x

xrange = range(-10,11)


for i in range(0,len(xrange)):
  plt.plot(xrange[i],y.subs({x:xrange[i]}),'o')

plt.xlabel('x')
plt.ylabel('$f(x) = %s$' %sym.latex(y))
plt.show()
```
