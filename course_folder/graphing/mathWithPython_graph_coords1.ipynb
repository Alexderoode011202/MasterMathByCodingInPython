```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

# basic plotting a red dot
plt.plot(x,y,'ro')

# set axis limits
plt.axis('square') # order matters
plt.axis([-6,6,-6,6])
plt.grid()

plt.show()
```

```python
# a set of coordinates

x = [-4,2,5,6,2,-5]
y = [5,2,10,-5,4,0]

for i in range(0,len(x)):
  plt.plot(x[i],y[i],'o',label='point %s'%i)


plt.legend()
plt.axis('square')
plt.grid()
plt.show()
```

```python
# getting information from axes

plt.plot(4,3,'rs')

# get an object for the current axis
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)

# now change only the upper y-axis limit
axis.set_ylim([ ylim[0],6 ])

plt.xlabel('X axis')
plt.ylabel('F(x)')

plt.show()
```

```python
import numpy as np
```

```python
x = sym.Symbol("x")
func = x**2-3*x

answers_dict: dict = {}
for index in range(-10, 12):
    answers_dict[index] = func.subs({x:index})
    
plt.scatter(list(answers_dict.keys()), list(answers_dict.values()))
plt.grid()
plt.axis("square")
plt.show()

random_points = np.random.randint(low=0, high=20, size=(20,2))
plt.scatter(random_points[:,0], random_points[:,1], color="green")
```

```python
random_points1 = np.random.randint(low=0, high=20, size=(20,2))
random_points2 = np.random.randint(low=0, high=20, size=(20,2))

plt.scatter(random_points1[:,0], random_points1[:,1],color=(0.5,0,0))
plt.plot(random_points2[:,0], random_points2[:,1], color="red")
plt.axis("square")
plt.grid()

plt.show()
```
