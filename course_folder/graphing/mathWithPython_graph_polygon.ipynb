```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
from matplotlib.patches import Polygon

x = np.linspace(0,1,100)
y = np.array([ [1,1],[2,3],[3,1] ])
p = Polygon(y,facecolor='m',alpha=.3)

# extend with two polygons
y1 = np.array([ [2,2],[2.5,4],[3.5,1] ])
p1 = Polygon(y1,alpha=.2,edgecolor='k')

fig, ax = plt.subplots()
ax.add_patch(p1)
ax.add_patch(p)
ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```

```python
x = np.linspace(-2,2,101)
f = -x**2

y = np.vstack((x,f)).T
p = Polygon(y,facecolor='g',alpha=.2,edgecolor='k')

p1 = Polygon(np.array([ [-.5,-4],[-.5,-2.5],[.5,-2.5],[.5,-4] ]),facecolor='k')

fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f,'k')
plt.plot(x[[0,-1]],[-4,-4],'k')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Access and modify individual subplots
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 0].bar([1, 2, 3], [4, 5, 6])
axs[1, 1].imshow([[1, 2], [3, 4]])

# Customize the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a new figure
fig = plt.figure()

# Add a subplot to the figure
ax1 = fig.add_subplot(2, 1, 1)  # 2 rows, 1 column, subplot 1
ax2 = fig.add_subplot(2, 1, 2)  # 2 rows, 1 column, subplot 2

# Plot data on the subplots
ax1.plot([1, 2, 3], [4, 5, 6])
ax2.scatter([1, 2, 3], [4, 5, 6])

# Customize figure properties
fig.suptitle('Example Figure')  # Overall title for the figure

# Show the plot
plt.show()
```

```python
from matplotlib.patches import Polygon
x_axis = np.linspace(-2, 2, 101)
y_axis = -x**2
pol1 = Polygon(tuple(zip(x_axis, y_axis)), facecolor='g',alpha=.2,edgecolor='Black')
pol2 = Polygon(((-0.5, -4), (0.5, -4), (0.5, -2.5), (-0.5, -2.5)), facecolor="Black")
fig, ax = plt.subplots()

ax.add_patch(pol1)
ax.add_patch(pol2)
ax.plot(x_axis, y_axis, color="Black")
ax.set_xlim(-2, 2)
ax.set_ylim(-4, 0.1)
ax.set_axis_off()
plt.show()
```
