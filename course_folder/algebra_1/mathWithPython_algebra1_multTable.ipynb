# Exercise: Create a multiplication table

```python
nums = range(1,11)

multmat = np.zeros((len(nums),len(nums)),dtype=int)

for i in nums:
  for j in nums:
    multmat[i-1,j-1] = i*j


display(Math(sym.latex(sym.sympify(multmat)))) # no display without display

x = 3
```
