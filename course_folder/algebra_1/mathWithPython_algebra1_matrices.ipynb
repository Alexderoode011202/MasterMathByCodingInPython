```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
sym.init_printing()
```

```python
A = np.array( [ [1,2],[3,4] ] )
print(A)

# make it look nicer
display(Math(sym.latex(sym.sympify(A))))

display(Math(sym.latex(sym.sympify(np.array([[1,0],[0,1]])))))
test = np.matrix(data=[[1,2],[3,4]])
display(sym.sympify(test))

print(sym.latex(sym.sympify(test)))

display(sym.sympify(3+test))
```

```python
# initializing a matrix with zeros

numrange = range(0,5)

mat = np.zeros([len(numrange),len(numrange)])
print(mat)
```

```python
# populating matrices using row-col indexing

mat[0,1] = 1
# mat[5,8] = 4
mat
```

```python
# can also use variables for indices
i = 2
j = 1
mat[i,j] = 4.5

display(Math(sym.latex(sym.sympify(mat))))
```

```python
# now use a for-loop

numrange = range(0,3)
for i in numrange:
  for j in numrange:
    mat[i][j] = (-1)**(i+j)

mat
```

```python
x,y = sym.symbols('x y')
Fxy = (4+x)*(2-y)

numrange = range(0,3)

funout = np.zeros((len(numrange),len(numrange)))

for i in numrange:
  for j in numrange:
    funout[i,j] = Fxy.subs({x:i,y:j})

display(Math(sym.latex(sym.sympify(funout))))
```

```python
x,y = sym.symbols("x, y")

f = (4+x)*(2-y)

matrix = np.zeros([3,3])
for column in range(0,3):
    for row in range(0,3):
        matrix[row, column] = f.subs({x: row, y: column})

display(sym.sympify(matrix))

```
